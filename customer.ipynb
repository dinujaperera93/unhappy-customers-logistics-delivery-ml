# Import Dependancies

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

# Load Dataset

```python
df_customer = pd.read_csv("ACME-HappinessSurvey2020.csv")
pd.concat([df_customer.head(5), df_customer.tail(5)])
```

```python
for col in df_customer.loc[ : , df_customer.columns != 'Y']:
    plt.plot(2,3)
    sns.countplot(data = df_customer, x=col, hue="Y")    
    plt.title(f'Distribution of {col}')
    plt.xlabel(f'{col}')
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()
```

```python
X = df_customer[df_customer.columns[~df_customer.columns.isin(["Y"])]]
y = df_customer["Y"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```
